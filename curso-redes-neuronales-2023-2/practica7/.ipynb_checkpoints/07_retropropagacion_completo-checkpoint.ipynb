# Entrenamiento con retropropagación

En este ejercicio implementaremos el algoritmo de retropropagación dentro del descenso por gradiente para actualizar todos los pesos de la red durante varias épocas. Para entrenar la red usaremos el conjunto de datos de calificaciones que vimos previamente.



```python
#importamos paquetes y datos
import numpy as np
from data_prep import features, targets, features_test, targets_test

# Definiciones útiles
np.random.seed(21)

def sigmoid(x):
    """
    Calculate sigmoid
    """
    return 1 / (1 + np.exp(-x))

def sigmoid_prime(x):
    return sigmoid(x)*(1-sigmoid(x))

```

```python
# Hyperparámetros
n_hidden = 2  # number of hidden units
epochs = 900
learnrate = 0.005

# Obtenemos el número de entradas (features) asi como el número de ejemplos (n_records)
n_records, n_features = features.shape
last_loss = None

# Creamos las matrices de los pesos.
weights_input_hidden = np.random.normal(scale=1 / n_features ** .5,
                                        size=(n_features, n_hidden))
weights_hidden_output = np.random.normal(scale=1 / n_features ** .5,
                                         size=n_hidden)
```

## Entrenamiento

```python

for e in range(epochs):
    del_w_input_hidden = np.zeros(weights_input_hidden.shape)
    del_w_hidden_output = np.zeros(weights_hidden_output.shape)
    for x, y in zip(features.values, targets):
        ## Forward pass ##
        # TODO: Calculate the output
        hidden_input = np.dot(x, weights_input_hidden)
        hidden_output = []
        
        for input in hidden_input:
            hidden_output.append(sigmoid(input))
            
        output_prev = np.dot(hidden_output, weights_hidden_output)
            
        output = sigmoid(output_prev) #el vector de pesos weights_hidden_output es unidimencsional, al igual que el vector de entradas que obtuvimos

        ## Backward pass ##
        # TODO: Calculate the error
        error = y - output

        # TODO: Calculate error gradient in output unit
        #imagino que esto es el termino de error
        
        output_error = error * sigmoid_prime(output_prev)

        # TODO: propagate errors to hidden layer
        
        hidden_error = []
        for neuron, weight in zip(hidden_input, weights_hidden_output):
             hidden_error.append(output_error * sigmoid_prime(neuron) * weight)

        # TODO: Update the change in weights
        del_w_hidden_output = []
        
        for neuron, weight in zip(hidden_input, weights_hidden_output):   #<- vamos en esta parte, aun no hemos terminado
            del_w_hidden_output.append(learnrate * output_error * weight)
            
        del_w_input_hidden = []
        
        for neuron in x:
            for error in hidden_error:
                del_w_input_hidden.append(learnrate * error * neuron)

    # TODO: Update weights
    #weights_input_hidden += del_w_input_hidden
    print(weights_input_hidden.shape[0])
    for i in range(weights_input_hidden.shape[0]):
        for j in range(weights_input_hidden.shape[1]):
            weights_input_hidden[i][j] += del_w_input_hidden[i]
    
    #weights_hidden_output += del_w_hidden_output
    
    for i in range(weights_hidden_output.shape[0]):
        weights_hidden_output[i] += del_w_hidden_output[i]

    # Printing out the mean square error on the training set
    if e % (epochs / 10) == 0:
        hidden_output = sigmoid(np.dot(x, weights_input_hidden))
        out = sigmoid(np.dot(hidden_output,
                             weights_hidden_output))
        loss = np.mean((out - targets) ** 2)

        if last_loss and last_loss < loss:
            print("Train loss: ", loss, "  WARNING - Loss Increasing")
        else:
            print("Train loss: ", loss)
        last_loss = loss

# Calculate accuracy on test data
hidden = sigmoid(np.dot(features_test, weights_input_hidden))
out = sigmoid(np.dot(hidden, weights_hidden_output))
predictions = out > 0.5
accuracy = np.mean(predictions == targets_test)
print("Prediction accuracy: {:.3f}".format(accuracy))
```
