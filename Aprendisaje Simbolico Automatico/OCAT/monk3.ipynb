```python
import pandas as pd
```

## importamos los dataset

```python
train_set = pd.read_excel("src/monks problems/monks-3-train.ods")
test_set = pd.read_excel("src/monks problems/monks-3-test.ods")

print(train_set)
print(test_set)
```

## Binarizar las características a1-a6

Recordemos que se trata de datos categoricos, por lo que, independientemente de que estén etiquetados con valores numeros, hay que darles el tratamiento de datos categoricos

```python
from Binarizacion import binarizar_categoricos

binarizados_dataframe:list[pd.DataFrame] = []

for i in range(1,7):
    binarizados_dataframe.append(binarizar_categoricos(train_set[f"a{i}"].values.tolist(), i))

# for binarizado in binarizados_dataframe:
#     print(binarizado)

#trampita, primero vamos a meter en forma de dataframe las clases a las que pertenece cada registro
binarizados_dataframe.append(train_set["Class"])

dataframe_binario = pd.concat(binarizados_dataframe, axis=1)

print(dataframe_binario)
```

## Una vez binarizado el dataset, separemos los registros por clase

```python
dataframe_binario_clase_0 = dataframe_binario[dataframe_binario["Class"] == 0]
dataframe_binario_clase_1 = dataframe_binario[dataframe_binario["Class"] == 1]

print(dataframe_binario_clase_0)
print(dataframe_binario_clase_1)
```

## comencemos a realizar las observaciones a lo largo de todo el dataset para encontrar cláusulas de preferencia completas y consistentes

```python
etiquetas_dataframe_binario = dataframe_binario.columns.to_list()

for i in range(1, 7):
    if etiquetas_dataframe_binario.__contains__(f"valor_real {i}"):
        etiquetas_dataframe_binario.remove(f"valor_real {i}")

etiquetas_dataframe_binario.remove("Class")

print(etiquetas_dataframe_binario)
```

```python
import itertools

combinaciones:list[list[str]] = []

for i in range(1, len(etiquetas_dataframe_binario)+1):

    for combinacion in itertools.combinations(etiquetas_dataframe_binario, i):
        combinaciones.append(list(combinacion))

print(len(combinaciones))
print(combinaciones[:20])
```

### Con las combinaciones, ahora podemos verificar las coberturas de todos los distintos complejors que podemos formar

```python
# cobertura:list[tuple[list[str], int]] = []

coberturas:list[int] = []
anticoberturas:list[int] = []

for complejo in combinaciones:
    sentencia:pd.Series = (dataframe_binario_clase_1[complejo[0]] == 1)
    sentencia2:pd.Series = (dataframe_binario_clase_0[complejo[0]] == 1)
    for i, elemento in enumerate(complejo):
        if i == 0:
            continue
        sentencia = sentencia & (dataframe_binario_clase_1[elemento] == 1)
        sentencia2 = sentencia2 & (dataframe_binario_clase_0[elemento] == 1)
    # print(sentencia)
    # print (f"{dataframe_binario_clase_1[sentencia].shape[0]} \t {dataframe_binario_clase_0[sentencia2].shape[0]}")   #con esto tenemos la cobertura!!!
    coberturas.append(dataframe_binario_clase_1[sentencia].shape[0])
    anticoberturas.append(dataframe_binario_clase_0[sentencia2].shape[0])
```

```python
combinaciones_consistentes:list[list[str]] = []

for i, cobertura in enumerate(coberturas):

    if cobertura > 0 and anticoberturas[i] == 0:
        combinaciones_consistentes.append(combinaciones[i])

print(combinaciones_consistentes)
```

### Comprobamos la cobertura total en el dataset

```python
from librerias_caseras.herramientas_listas import list_to_series_disyuntivas

print(dataframe_binario_clase_1[list_to_series_disyuntivas(combinaciones_consistentes, dataframe_binario_clase_1)])
```
